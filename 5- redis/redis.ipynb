```python
import redis
import pandas as pd
```

```python
#import data
dataFrame = pd.read_csv('titanic.csv')
dataFrame.head()
```

```python
# show data
dataFrame['Name']
```

```python
#create redis driver instance
r = redis.Redis()

# add into redis with sadd
for index, value in dataFrame.iterrows():
    r.sadd(value['Name'], *value)
```

```python
#get from redis with smembers
for index, value in dataFrame.iterrows():
    print(r.smembers(value['Name']))
```
